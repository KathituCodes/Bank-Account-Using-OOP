```python
import random
import string

# Step 1 Create a Class called "Account"
class Account:
    def __init__(self, account_number, account_balance, account_holder):
        self.account_number = account_number
        self.account_balance = account_balance
        self.account_holder = account_holder

    """
    Define the deposit() method. It should take in one argument,
    the amount to be deposited, and add it to the account balance.
    """
    def deposit(self):
        deposit = float(input("Enter the amount that you would wish to deposit: "))
        self.account_balance += deposit
        print(f"You've successfully deposited {deposit} ksh.")

    """"
    Define the withdraw() method. It should take in one argument,
    the amount to be withdrawn, and subtract it from the account balance.
    The method should only execute the withdrawal if the account balance
    is greater than or equal to the amount to be withdrawn.
    """
    def withdraw(self):
        withdraw = float(input("Enter the amount that you would wish to withdraw: "))
        if withdraw <= self.account_balance:
            self.account_balance -= withdraw
            print(f"You've successfully withdrawn {withdraw} ksh.")
        else:
            print("You have insufficient balance.")

    """"
    Define the check_balance() method.
    It should return the current account balance.
    """
    def check_balance(self):
        print(f"Your account balance is {self.account_balance} ksh.")


    def __str__(self):
        return f"Account Holder: {self.account_holder}, Account Number: {self.account_number}, Balance: {self.account_balance} ksh"


    # Generate a random alphanumeric string of length 8
    def generate_account_number(self, account_holder: str) -> str:
        random_string = ''.join(random.choices(string.ascii_uppercase + string.digits, k=8))
        return random_string

# Step 2 Create an instance of the Account class

# Create an empty dictionary that stores the acoount details
accounts = {}

# Print Intro and Instructions
print("Welcome to Standard Chartered Bank!")
print("Follow the instructions below to Create or Manage an Account")

while True:
    print("""
    MENU
          1. To create an account
          2. To manage the account
          3. Exit
""")
    choice = input("Enter an option (1, 2, or 3 to exit): ")  # Keep choice as a string initially

    # Validate user input before converting to int
    if choice.isdigit() and choice in ('1', '2', '3'):
        choice = int(choice)
    else:
        print("Invalid input! Please enter 1, 2, or 3.")
        continue # Go back to the beginning of the loop

    if choice == 1:
        account_holder = input("Enter account holder name: ")
        account_number = Account("", 0, "").generate_account_number(account_holder)
        accounts[account_holder] = Account(account_number, 0, account_holder)
        print(f"Account created for {account_holder} with account number {account_number}")

    elif choice == 2:
        if not accounts:  #check if the account exists
            print("Please create an account first.")
        else:
            for holder, account in accounts.items():
                print(account)
            acc = input("Select an account to manage: ")
            if acc in accounts:
                selected_acc = accounts[acc]
                while True:
                    print("""
                        ACCOUNT OPERATIONS:
                        -------------------
                            1. To deposit
                            2. To withdraw
                            3. To check balance
                            4. To go back to the main menu
                    """)
                    option = input("Select an account operation: ")
                    if option.isdigit() and option in ('1', '2', '3', '4'):
                        option = int(option)
                    else:
                        print("Invalid input! Please enter 1, 2, 3, or 4.")
                        continue
                    if option == 1:
                        selected_acc.deposit()
                    elif option == 2:
                        selected_acc.withdraw()
                    elif option == 3:
                        selected_acc.check_balance()
                    elif option == 4:
                        break
            else:
                print("Invalid account holder.")

    elif choice == 3:
        break
```
